```python
from tkinter import *
from tkinter import messagebox
import numpy as np
```

```python
class connect4_gui:
  def __init__(self, color = None):
    self.finish = 0
    self.root = Tk()
    self.root.title("Connect4")
    self.turn = 'blue'
    if color:
      self.turn = color
    self.buttons = np.empty((6, 7), dtype=object)
    self.levels = np.zeros(7, dtype='int32')
    frame = Frame(self.root, bg='#6AECE1', padx= 10, pady=10)
    for row in range(6):
      for col in range(7):
        button = Button(frame, padx=40, pady=40, bg='#435663', command=lambda c =col: self.check_turn_and_update(c))
        button.grid(row=row, column=col, padx=1, pady=1)
        self.buttons[row, col] = button
    frame.pack(padx=20, pady=20)
    self.root.mainloop()
  def check_turn_and_update(self, column):
    if self.levels[column] > 5:
      messagebox.showinfo("Error messege", "This column is filled")
      return
    self.buttons[5-self.levels[column], column].config(bg = self.turn)
    self.levels[column] += 1
    self.check_end_game()
    if self.turn == 'blue':
      self.turn = 'red'
    else:
      self.turn = 'blue'
  def check_end_game(self):
    for row in range(6):
      for col in range(7-3):
        if self.buttons[row][col].cget('bg') == self.buttons[row][col+1].cget('bg') == self.buttons[row][col+2].cget('bg') == self.buttons[row][col+3].cget('bg') and self.buttons[row][col].cget('bg') != '#435663':
          messagebox.showinfo("Check winner", f'{self.turn} is the winner')
          self.root.destroy()
          return
    for row in range(6-3):
      for col in range(7):
        if self.buttons[row][col].cget('bg') == self.buttons[row+1][col].cget('bg') == self.buttons[row+2][col].cget('bg') == self.buttons[row+3][col].cget('bg') and self.buttons[row][col].cget('bg') != '#435663':
          messagebox.showinfo("Check winner", f'{self.turn} is the winner')
          self.root.destroy()
          return
    if np.all(self.levels == 6):
      messagebox.showinfo("Check winner", f'No one wins')
      self.root.destroy()
      return
```
